Headers were not includes in csv file but were available on the source website so we manually add headers so data can be a bit more meaningful

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

### Handling missing values
Since we are pridicting price rows without price is not useful to us.<br>
We need to replace the "?" symbol with NaN so the dropna() can remove the missing values

```python
df1=df.replace('?',np.NaN)
```

```python
df=df1.dropna(subset=["price"], axis=0)
df.head()
```

### How to handle missing values
1. drop data
2. replace data based on function if available
3. replace data with either mean or mode 

we can find missing values using <code>isnull(), notnull()</code> methods

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

Based on the summary above, each column has 205 rows of data and seven of the columns containing missing data:

<ol>
    <li>"normalized-losses": 41 missing data</li>
    <li>"num-of-doors": 2 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
    <li>"price": 4 missing data</li>
</ol>

Here we can change most of the values with their mean value we have to drop "price" as we are predicting it and cant replace it with mean and we have to substitute "number of doors" with mode as it can only be integer

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

For door we replace with mode, We can also use <code>idmax()</code> to find value with maximum occurance in this case "four"

```python
df['num-of-doors'].value_counts()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
df=df.dropna(subset=["price"], axis=0)
df.reset_index(drop=True, inplace=True)
```

Lets see our dataset now

```python
df.head()
```

<p>As we can see above, some columns are not of the correct data type. Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'. For example, 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. We have to convert data types into a proper format for each column using the "astype()" method.</p> 

<h4>Convert data types to proper format</h4>

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
